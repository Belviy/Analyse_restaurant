```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Commande=pd.read_csv("Commande_de_norriture.csv")
Commande.head()
```

```python
plt.figure(figsize=(10,5))
plt.hist(Commande['Taille Famille'],bins=12,color="Green")
plt.title("Taille de Famille a cible",fontsize=20, weight="bold")
plt.xlabel("Taille de Famille",fontsize=10,weight="bold")
plt.ylabel("Nombre de person",fontsize=10,weight="bold")

sns.despine()
plt.savefig("taillefamille.png")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(Commande['Age'],bins=12,color="Blue")
plt.title("Tranche d'age a cible",fontsize=20, weight="bold")
plt.xlabel("Age",fontsize=10,weight="bold")
plt.ylabel("Nombre de person",fontsize=10,weight="bold")

sns.despine()
plt.savefig("Age.png")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=Commande,x=Commande["Salaire Mensuelle"],hue='Occupation',palette="Reds")
plt.title("Salaire par occupation",fontsize=20,fontweight='bold')
plt.xlabel("Salaire",fontsize=15,fontweight='bold')
plt.ylabel("Nombre de personne",fontsize=9,fontweight='bold')
plt.legend(title='Occupation') 
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(Commande['Commentaire'],bins=3,color="#108A98")
plt.xlabel("Commentaire",fontsize= 8,fontweight='bold')
plt.title("Impact des Commentaires",fontsize= 15,fontweight='bold')
plt.ylabel("Nombre de person",fontsize= 8,fontweight='bold')
plt.xticks([0,1],['Negative','Positive'])
sns.despine()
plt.savefig("Commentaire.png")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(data=Commande,x=Commande['Genre'],hue='Status',palette="Blues_r")
plt.title("Status et Genre",fontsize=15,weight='bold')
plt.xlabel("Genre",fontsize=10,weight='bold')
plt.ylabel("Nombre de person",fontsize=10,weight='bold')  
plt.legend(labels=['Student','Employeed','House wife'],title='Status')
plt.xticks([0,1],['Female','Male'],rotation=45)
sns.despine()
plt.savefig("Genre.png")
plt.show()
```
