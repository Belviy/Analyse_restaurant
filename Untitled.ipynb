```python
import pandas as pd
import numpy as np
```

```python
#lire les donnees
data=pd.read_csv("onlinefoods.csv")
data.head()
```

```python
data.columns
```

```python
data.isna().sum()
```

```python
data=data.drop(['Educational Qualifications', 'latitude', 'longitude','Pin code', 'Output','Delivered'],axis=1)
data
```

```python
data=data.rename(columns={'Gender':'Genre','Marital Status':'Status','Monthly Income':'Salaire Mensuelle'
                         ,'Family size':'Taille Famille','Feedback':'Commentaire'})
data
```

```python
data.drop_duplicates(inplace=True,keep='first')
data
```

```python
data['Genre']=np.where(data['Genre']=='Female','0',data['Genre'])
data['Genre']=np.where(data['Genre']=='Male','1',data['Genre'])
data['Status']=np.where(data['Status']=='Single','0',data['Status'])
data['Status']=np.where(data['Status']=='Married','1',data['Status'])

data['Status']=np.where(data['Status']=='Prefer not to say','2',data['Status'])

data['Commentaire']=np.where(data['Commentaire']=='Negative ','0',data['Commentaire'])
data['Commentaire']=np.where(data['Commentaire']=='Positive','1',data['Commentaire'])
data
```

```python
data['Commentaire'].unique()
```

```python
data['Occupation'].unique()
```

```python
data['Occupation']=np.where(data['Occupation']=='Self Employeed','Employeed',data['Occupation'])
data['Occupation']=np.where(data['Occupation']=='Employee','Employeed',data['Occupation'])
data
```

```python
data['Salaire Mensuelle'].unique()
```

```python
data['Salaire Mensuelle']=np.where(data['Salaire Mensuelle']=='No Income','0$',data['Salaire Mensuelle'])
data['Salaire Mensuelle']=np.where(data['Salaire Mensuelle']=='Below Rs.10000','1-10000$',data['Salaire Mensuelle'])

data['Salaire Mensuelle']=np.where(data['Salaire Mensuelle']=='10001 to 25000','10001-25000$',data['Salaire Mensuelle'])
data['Salaire Mensuelle']=np.where(data['Salaire Mensuelle']=='25001 to 50000','25001-50000$',data['Salaire Mensuelle'])

data['Salaire Mensuelle']=np.where(data['Salaire Mensuelle']=='More than 50000','50000-100000$',data['Salaire Mensuelle'])
data
```

```python
data.to_csv("Commande_de_norriture.csv",index=False)
```
